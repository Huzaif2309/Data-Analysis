```python
selected_columns = ['Delhi','Tamil Nadu', 'Karnataka', 'Maharashtra']


df = pd.read_csv('C:/Users/admin/Downloads/TB2.csv', usecols=selected_columns)

df = df.applymap(str)

selected_rows = [6,12,18]
for row_index in selected_rows:
    if 0 < row_index <= len(df):
        plt.plot(df.columns, df.iloc[row_index - 1], label=f'Row {row_index}', marker='o')
    else:
        print(f"Row {row_index} is out of bounds for the DataFrame.")

plt.title('')
plt.xlabel('Metropolitain States')
plt.ylabel('Percentages')
plt.grid(True)
plt.legend()

plt.show()
```

```python
selected_columns = ['Arunachal Pradesh', 'Jharkhand', 'Bihar','Rajasthan']

df = pd.read_csv('C:/Users/admin/Downloads/TB2.csv', usecols=selected_columns)

df = df.apply(pd.to_numeric, errors='coerce')

selected_rows = [6, 18,12]
bar_width = 0.25  
bar_colors = ['orange', 'green', 'blue']

for i, row_index in enumerate(selected_rows):
    if 0 < row_index <= len(df):
        bars = plt.bar(np.arange(len(df.columns)) + i * bar_width, df.iloc[row_index - 1],
                       width=bar_width, label=f'Row {row_index}', color=bar_colors[i])

plt.title('')
plt.xlabel('Low Literacy Rate States')
plt.ylabel('Percentages')
plt.xticks(np.arange(len(df.columns)) + bar_width, df.columns)  # Set x-axis ticks
plt.legend()
plt.show()
```

```python
selected_columns = ['Odisha', 'Kerala', 'Bihar']

df = pd.read_csv('C:/Users/admin/Downloads/TB2.csv', usecols=selected_columns)

df = df.applymap(str)

selected_rows = [6, 12, 18]

percentages_list = []

for row_index in selected_rows:
    if 0 < row_index <= len(df):
        percentages_list.append(df.iloc[row_index - 1].astype(float))
    else:
        print(f"Row {row_index} is out of bounds for the DataFrame.")

for i, percentages in enumerate(percentages_list):
    plt.figure()
    plt.pie(percentages, labels=df.columns, autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart for Row {selected_rows[i]}')

plt.show()
```
